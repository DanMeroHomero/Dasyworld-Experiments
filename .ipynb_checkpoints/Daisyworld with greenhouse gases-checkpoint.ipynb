# Daisyworld with greenhouse gases

In this notebook we will study a model that introduces greenhouse gases into Daisyworld. This model was created by Paiva et. al (2014). These authors modified the growth rate of the daisies, and the equation of radiative balance of the original Daisyworld model, to study global warming in this planet. 

First, they define the growth rate of the daisies as a gaussian function with a maximum growth rate at $22.5°C$:

\begin{equation}
\beta (T_i) = B e^{-\frac{1}{s} (T_i - T_{opt})^2}
\end{equation}

Where $B$ is a constant related to environmental factors, $s$ is the variability of the distribution, $T_i$ is the temperature of the daisies, and $T_{opt} = 22.5°C$ is the growth optimum temperature . Let's plot this function to see how it looks, and how the parameters $B$ and $s$ affect the shape of the distribution. 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def growth_rate(B, s):
    T = np.arange(5, 40, 0.2)
    T_opt = 22.5
    beta = B * np.exp(-(T - T_opt)**2 / s)
    return beta

T = np.arange(5, 40, 0.2)
fig, axs = plt.subplots(1, 3, figsize=(15, 3))
growth_rates = {0:[growth_rate(1, 150),1, 150],
               1:[growth_rate(0.5, 150), 0.5, 150],
               2:[growth_rate(1, 10), 1, 10]}
for i, growth_rate in growth_rates.items():
    axs[i].plot(T, growth_rate[0], linewidth = 2.5)
    axs[i].set_xlabel('Temperature °C')
    axs[i].set_ylabel('growth rate')
    axs[i].axis([5, 40, 0, 1.02])  
    axs[i].set_title(f'B = {growth_rate[1]}, s = {growth_rate[2]}', fontsize = 17)
```

In the above cell, we see that when $B=1$ and $s=150$, we obtain a normal distribution of temperature that achieves a maximum growth rate of $1$ at $22.5^{\circ}C$ and that decays near $5^{\circ}C$ and $40^{\circ}C$. In contrast, when we set $B=0.5$ and $s=150$, the maximum growth rate is 0.5. That is why the authors mention that the $B$ parameter is related to environmental conditions, since $B$ controls the maximum growth rate of the daisies. In addition, when we set $B=1$ and $s=10$ we obtain a thiner distribution. This is because $s$ controls the variability from the optimum temperature. Here we will be using $B=1$ and $s=150$. The authors say that using a gaussian function, rather than a parable like in the original model, prevents the daisies from reducing their population in an abrupt way when the temperature is near $5^{\circ}C$ or $40^{\circ}C$.

Now let's see how Paiva etl al. (2014) modified the radiative balance of Daisyworld. Following Nevison et. al. (1999), they introduced a ODE to model the change in the effective temperature due to the radiative unbalance, but this time they added a term for the green house effect:

\begin{equation} 
\frac{dT_e}{dt} = \frac{1}{c}[S L(1 -A) - \sigma T_{e}^4(1-E)]
\end{equation}

Where $T_e$ is the effective temperature of Daisyworld, $c$ is the thermal inertia of the planet, $S$ is the radiation reaching the planet, $L$ is a dimensionless number that controls the solar luminosity, $A$ is the average albedo of the planet, $\sigma$ is the Steffan-Botlzman's constant, and $E$ is a term that represents the green house effect, and is given by the equation:

\begin{equation}
\frac{d E}{dt} = a - b E (\alpha_w + \alpha_b)
\end{equation}

Where $a$ is a parameter related to the emission of green house gases, $b$ is related to the gas absorption, and $\alpha_w$ and $\alpha_b$ are the fraction areas of white and black daisies in the planet.

Let's analyze this equations quantitatively. The equation for the effective temperature depends on the total radiation reaching the planet, and on the total radiation that the planet emits. If the radiation reaching the planet is greater than the radiation leaving the planet, the effective temperature will augment at a rate controlled by the thermal inertia. And in the opposite case, the temperature will augment. Now, the term for the green house effect $E$ affects the total radiation that the planet emits. If it is negative, it augments the emitted radiation and decreases the temperature. If it is positive, it decreases the emitted radiation and makes the temperature increase.

Now, $E$ depends on the term for emission of gases $a$ and on the term for absorption $b$. If the emission of gases in greater than the absorption, $E$ will be positive and it will contribute to augment the effective temperature. In the opposite case it will decrease the temperature. When there is a perfect balance, $E=0$ and there is no green house effect.

Now that we have a qualitative understanding of this equations, let's solve the model to see how introducing the greenhouse effect in Daisyworld affects the daisies. We will use the same code structure as in the Original Daisyworld model, adding the necessary changes in the growth rate and in the equation for radiative balance.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import e
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

class Greenhouse_Daisywolrd():
    q = 2064250710.0 #20 # Heat Transfer Coefficient
    sigma = 49e-3 #1789  #Stefan-Boltzman Constant
    S = 79.1e+7 #2.89e+13  # Solar constant
    
    
    def __init__(self
                ,death_rate =.3 #death rate of daysies
                ,albedo_w=0.75  #white daysies albedo
                ,albedo_g=0.5  #Free ground albedo      
                ,albedo_b=0.25  #Black daysies albedo
                ,area_w= 0.3 # Fraction area of white daisies
                ,area_b= 0.3 # Fraction area of black daisies
                ,cp = 82.1e+7 #3e+13 # thermal inertia
                ,L = 1 # Solar luminosity constant
                ,a = 1e-4 # Green house emission constant
                ,b = 1e-1 # Green house absorption constant
                ):
        'Initilize all the constants and variables'
        self.death_rate = death_rate
        self.albedo_w = albedo_w
        self.albedo_g = albedo_g
        self.albedo_b = albedo_b
        self.area_w = area_w
        self.area_b = area_b
        self.area_g = 1 - self.area_b - self.area_w # Equation 2
        self.cp = cp
        self.L = L
        self.a = a
        self.b = b
        self.albedo_total = self.area_w * self.albedo_w + self.area_b * self.albedo_b + self.area_g * self.albedo_g
        self.E = 0
        self.Te = ((self.S * self.L * (1- self.albedo_total) / self.sigma) ** 0.25)- 273 #Equation 4
        self.Tw = (self.q*(self.albedo_total - self.albedo_w) + (self.Te +273)**4 )**.25 -273
        self.Tb = (self.q*(self.albedo_total - self.albedo_b)+ (self.Te +273)**4 )**.25 -273
        self.Bw = 1 -0.003265 * ((22.5-self.Tw)**2) #Equation 2
        self.Bb = 1 -0.003265 * ((22.5-self.Tb)**2) #Equation 2
        
    def update_area_g(self):
        '''Calclates area of free ground'''
        self.area_g = 1 - self.area_b - self.area_w 
        return self.area_g
    
    def update_albedo_total(self):
        '''Calculates average albedo of Dasyworld'''
        self.albedo_total = self.area_w * self.albedo_w + self.area_b * self.albedo_b + self.area_g * self.albedo_g

    def update_green_house_effect(self, h):
        '''Solves greenhouse effect equation with RK4'''
        kE1 = h* (self.a - self.b * self.E * (self.area_w + self.area_b))
        kE2 = h * (self.a - self.b * (self.E + kE1 /2.0) * (self.area_w + self.area_b))
        kE3 = h * (self.a - self.b * (self.E + kE2 /2.0) * (self.area_w + self.area_b))
        kE4 = h * (self.a - self.b * (self.E + kE3) * (self.area_w + self.area_b))
        self.E += (kE1 + 2 * kE2 + 2 * kE3 + kE4 ) / 6.0
        
    def update_global_temp(self, h, L):
        '''Solves effective temperature equation with RG4'''
        kT1 = h * (self.S * L * (1- self.albedo_total) - \
                        self.sigma * (1 - self.E) * ((self.Te + 273) **4) ) / self.cp 
        kT2 = h * (self.S * L * (1- self.albedo_total) - \
                        self.sigma * (1 - self.E) * (((self.Te + kT1/2 )  + 273) **4) ) / self.cp
        kT3 = h * (self.S * L * (1- self.albedo_total) - \
                        self.sigma * (1 - self.E) * (((self.Te + kT2/2 )  + 273) **4) ) / self.cp
        kT4 = h * (self.S * L * (1- self.albedo_total) - \
                        self.sigma * (1 - self.E) * (((self.Te + kT3)  + 273) **4) ) / self.cp
        self.Te += (kT1 + 2 * kT2 + 2 * kT3 + kT4 ) / 6.0

    
    def update_daisies_temp(self):
        '''Solves daisies temperature equation with RK4'''
        self.Tw = (self.q*(self.albedo_total - self.albedo_w) + (self.Te +273)**4 )**.25 -273
        self.Tb = (self.q*(self.albedo_total - self.albedo_b)+ (self.Te +273)**4 )**.25 -273
        #Linear aproximation case (q = 20)
        #self.Tw = self.q*(self.albedo_total-self.albedo_w)+ self.Te 
        #self.Tb = self.q*(self.albedo_total-self.albedo_b)+ self.Te 
    
    def update_growth_rate(self):
        '''updates growth rate of daisies'''
        self.Bw = e ** (- (self.Tw - 22.5)**2 / 150)
        self.Bb = e ** (- (self.Tb - 22.5)**2 / 150)
    
    def update_daisies_areas(self, h):
        '''Solves equation for daisies areas with RK4'''
        kw1 = h * self.area_w * (self.Bw * self.area_g -self.death_rate)
        kw2 = h * (self.area_w + kw1/2) * (self.Bw * self.area_g -self.death_rate)
        kw3 = h * (self.area_w + kw2/2) * (self.Bw * self.area_g -self.death_rate)
        kw4 = h * (self.area_w + kw3) * (self.Bw * self.area_g -self.death_rate)
        self.area_w += (kw1 + 2 * kw2 + 2 * kw3 + kw4 ) / 6
        kb1 = h * self.area_b * (self.Bb * self.area_g -self.death_rate)
        kb2 = h * (self.area_b + kb1/2) * (self.Bb * self.area_g -self.death_rate)
        kb3 = h * (self.area_b + kb2/2) * (self.Bb * self.area_g -self.death_rate)
        kb4 = h * (self.area_b + kb3) * (self.Bb * self.area_g -self.death_rate)
        self.area_b += (kb1 + 2 * kb2 + 2 * kb3 + kb4 ) / 6.0
    
    
    def solve_model(self, iterations = 12000, h = 0.1, store_values = False, L= None):
        '''Solves model calculating all variables in each step'''
        time = np.arange(0, iterations * h, h)
        AW, AB, AG, TW, TB, TE, E = ([],[],[],[],[],[],[])
        for i in range(iterations):
            #self.area_w += h * self.area_w * (self.Bw * self.area_g -self.death_rate)
            #self.area_b += h * self.area_b * (self.Bb * self.area_g -self.death_rate)
            self.update_green_house_effect(h)
            if isinstance(L, (np.ndarray, list, tuple)):
                self.update_global_temp(h, L[i])
            else:
                self.update_global_temp(h, self.L)
            self.update_daisies_areas(h)
            self.update_area_g()
            self.update_albedo_total()
            self.update_daisies_temp()
            self.update_growth_rate()
            if store_values:
                AW.append(self.area_w)
                AB.append(self.area_b)
                AG.append(self.area_g)
                TW.append(self.Tw)
                TB.append(self.Tb)
                TE.append(self.Te) 
                E.append(self.E)
        if store_values:
            self.AW = AW
            self.AB = AB
            self.AG = AG
            self.TW = TW
            self.TB = TB
            self.TE = TE
            self.E = E
            self.time = time
    
    def plot_model(self, Tw = False, Tb = False, Te = True, limits = True):
        '''Makes a plot of the daises areas and of the temperatures with time'''
        time = self.time
        fig, axs = plt.subplots(1, 2, figsize=(12, 3.5))
        axs[0].plot(time, self.AB, 'k', linewidth=2, label='Black')
        axs[0].plot(time, self.AW, 'g',linewidth=2,label='White')
        axs[0].set_xlabel('Time (days)', fontsize=18)
        axs[0].set_ylabel(r'Fraction area area_i', fontsize=18)
        axs[0].set_title('Change in daisies population', fontsize=20)
        legend = axs[0].legend(loc=1, shadow=True)
        frame = legend.get_frame()
        frame.set_facecolor('0.90')
        
        if Tb:
            axs[1].plot(time, self.TB, 'k:', linewidth=2, label='Tb')
        if Tw:
            axs[1].plot(time, self.TW, 'g:', linewidth=2, label='Tw' )
        if Te:
            axs[1].plot(time, self.TE, 'k', linewidth=2, label='Te' )
        axs[1].set_xlabel('Time (days)', fontsize=18)
        axs[1].set_ylabel(r'Temperature (°C)', fontsize=18)
        axs[1].set_title('Change in Temperature', fontsize=20)
        if limits:
            axs[1].set_ylim(0, 40)
        axs2 = axs[1].twinx()
        axs2.plot(time, self.E, 'r:', linewidth=1, label='E' )
        axs2.set_ylabel('E', fontsize=20)
        if limits:
            axs2.set_ylim(0, 1)
        legend = axs[1].legend(loc=1, shadow=True)
        legend2 = axs2.legend(loc=2, shadow=True)
        frame = legend.get_frame()
        frame.set_facecolor('0.90')
        
        
    
    def change_luminosity(self, L, iterations = 12000, h = 0.1):
        '''
        Solves model for different values of Solar Luminosity L, storing the final values
        INPUT:
            L(list): list describing change in luminosity.
        '''
        #iteration parameter
        n = len(L)
        # lists to store values
        change_lum_TE, change_lum_AW, change_lum_AB, change_lum_AG = ([],[],[],[])
        # initial values
        area_w, area_b, albedo_w, albedo_b, albedo_g, death_rate, a, b =\
        self.area_w, self.area_b, self.albedo_w, self.albedo_b, self.albedo_g, self.death_rate, self.a, self.b
        for i in range(n):
            model = Greenhouse_Daisywolrd(area_w=area_w, area_b = area_b, a=a, b=b, L = L[i])
            #model.L = L[i]
            model.solve_model(iterations, h)
            change_lum_AW.append(model.area_w)
            change_lum_AB.append(model.area_b)
            change_lum_AG.append(model.area_g)
            change_lum_TE.append(model.Te)
        self.change_lum_TE = np.array(change_lum_TE)
        self.change_lum_AW = np.array(change_lum_AW)
        self.change_lum_AB = np.array(change_lum_AB)
        self.change_lum_AG= np.array(change_lum_AG)
        self.L = L
        self.no_life_Te = (self.S * self.L * (1- 0.5) / self.sigma)**0.25-273
    
    def plot_change_luminosity(self):
        '''Makes plots of the change in solar luminosity'''
        fig, axs = plt.subplots(1, 2, figsize=(12, 3.5))
        axs[0].plot(self.L, self.change_lum_AB, 'k', linewidth=3, label='Black')
        axs[0].plot(self.L, self.change_lum_AW, 'g',linewidth=3,label='White')
        axs[0].set_xlabel('Solar Luminosity $L$', fontsize=18)
        axs[0].set_ylabel(r'Fraction area area_i', fontsize=18)
        axs[0].set_title('Change in daisies population', fontsize=20)
        axs[0].axis([0.4, 1.8, -0.1, 0.8])
        legend = axs[0].legend(loc=1, shadow=True)
        frame = legend.get_frame()
        frame.set_facecolor('0.90')
        for label in legend.get_texts():
            label.set_fontsize('large')
        for label in legend.get_lines():
            label.set_linewidth(1.5)  
            
        axs[1].plot(self.L, self.change_lum_TE, 'g', linewidth=3, label='T with daisies')
        axs[1].plot(self.L, self.no_life_Te , 'k--', linewidth=3, label='T without daisies')
        axs[1].set_xlabel('Solar Luminosity $L$', fontsize=18)
        axs[1].set_ylabel('$T $ [$^\circ C$]', fontsize=18)
        axs[1].set_title('Change in effective Temparetaure', fontsize=20)
        axs[1].axis([0.4, 1.6, -30, 90])
        legend = axs[1].legend(loc=4, shadow=True)
        frame = legend.get_frame()
        frame.set_facecolor('0.90')
        for label in legend.get_texts():
            label.set_fontsize('large')
        for label in legend.get_lines():
            label.set_linewidth(1.5)  
        plt.show()
        
    def plot_state_space(self, zlim=(0,1)):
        '''Makes 3d plots of the daisies areas and the greenhouse effect'''
        fig = plt.figure(figsize=(14, 7))
        ax = fig.add_subplot(1, 2, 1, projection='3d')
        ax.plot(self.AW, self.AB, self.E, label='parametric curve')
        ax.view_init(30, 310)
        ax.set_xlim3d(0, 0.7)
        ax.set_ylim3d(0,0.7)
        ax.set_zlim3d(zlim[0],zlim[1])
        ax.set_xlabel('Area_w', fontsize = 14)
        ax.set_ylabel('Area_b', fontsize = 14)
        ax.set_zlabel('E', fontsize = 14)
        ax.set_title('3d plot', fontsize=14)
        ax.view_init(20, 312)
        plt.draw()

        ax = fig.add_subplot(1, 2, 2, projection='3d')
        null = [0]*len(self.AW)
        ax.plot(self.AW, self.AB, null)
        ax.plot(self.AW,[0.71]*len(self.AW), self.E)
        ax.plot(null,self.AB, self.E)
        ax.view_init(20, 312)
        ax.set_xlim3d(0, 0.7)
        ax.set_ylim3d(0,0.7)
        ax.set_zlim3d(zlim[0],zlim[1])
        ax.set_xlabel('Area_w', fontsize = 14)
        ax.set_ylabel('Area_b', fontsize = 14)
        ax.set_zlabel('E', fontsize = 14)
        ax.set_title('Projections', fontsize=14)
        plt.draw()
            
    def plot_average_change_luminosity(self, L):
        '''Solves model for different values of L and for different number of iterations and
        makes plot of the results'''
        iterations = np.arange(1000, 12000, 200)
        #effective temperature without life
        no_life_Te = (2.89e+13 * L * (1- 0.5) / 1789)**0.25-273
        #lists to store values
        Te_min,Te_max, Te_mean, area_b_mean, area_w_mean = [], [], [], [], []
        area_w, area_b, albedo_w, albedo_b, albedo_g, death_rate, a, b  =\
        self.area_w, self.area_b, self.albedo_w, self.albedo_b, self.albedo_g, self.death_rate, self.a, self.b
        #for each luminosity value
        for value in L:
            Te = []
            ab = []
            aw = []
            #run the model for different times
            for iteration in iterations:
                model = Greenhouse_Daisywolrd(L = value, area_w=area_w, area_b = area_b, a=a, b=b)
                model.solve_model( iterations = iteration)
                Te.append(model.Te)
                ab.append(model.area_b)
                aw.append(model.area_w)
            #Calculate average, min and max of variables
            Te_min.append(np.array(Te).min())
            Te_max.append(np.array(Te).max())
            Te_mean.append(np.array(Te).mean())  
            area_b_mean.append(np.array(ab).mean())
            area_w_mean.append(np.array(aw).mean())

        #Make plots
        fig, axs = plt.subplots(1,2, figsize = (12, 3.5))
        axs[0].plot(L, area_b_mean, 'k', linewidth=3, label='white')
        axs[0].plot(L, area_w_mean, 'g',linewidth=3, label='black')
        axs[0].plot(L, np.array(area_b_mean) + np.array(area_w_mean), 'k:' ,linewidth=2, label='sum')
        axs[0].set_xlabel('Solar Luminosity $L$', fontsize=18)
        axs[0].set_ylabel('Average area_i', fontsize=18)
        axs[0].set_title('Change in daisies population', fontsize=20)
        legend = axs[0].legend(loc=0, shadow=True)
        frame = legend.get_frame()
        frame.set_facecolor('0.90')
        axs[1].plot(L, Te_mean, 'g', linewidth=2, label='$T_{mean}$')
        axs[1].plot(L, Te_min, 'g--',linewidth=2, label='$T_{min}$')
        axs[1].plot(L, Te_max, 'g:' ,linewidth=2, label='$T_{max}$')
        axs[1].plot(L, no_life_Te, 'k:' ,linewidth=2, label='$T_{nolife}$')
        axs[1].set_xlabel('Solar Luminosity $L$', fontsize=18)
        axs[1].set_ylabel('Temperature (°C)', fontsize=18)
        axs[1].set_title('Change in effective Temparetaure', fontsize=20)
        legend = axs[1].legend(loc=0, shadow=True)
        frame = legend.get_frame()
        frame.set_facecolor('0.90')

        
```

First, let's see what happens when there is no greenhouse effect in Daisyworld. This occurs then $a=b=0$ and $E=0$. To solve the model in this case we just have to create a Greenhouse_Daisy world object with the $a$ and $b$ parameters set to zero, and then use the `solve_model` and `plot_model` methods.

```python
model = Greenhouse_Daisywolrd( a=0, b=0)
model.solve_model(store_values = True, iterations = 10000)
model.plot_model()
```

In this case we obtain a similar behavior to the one in the Daisyworld without radiative equilibrium. The populations of daisies and the effective temperature engage in coupled self-sustained oscillations. This makes sense because when we set $E=0$ we obtain the equation that describes radiative unbalance in the Daisyworld without radiative equilibrium. Paiva et. al. (2014) also plot the area of white daisies against the area of back daisies to see the state space of the daisies. In the following cell we can see this plot showing the oscillations of the daisy's populations. 

```python
plt.figure()
plt.plot(model.AW, model.AB, 'k')
plt.xlabel('Area_w', fontsize = 16)
plt.ylabel('Area_b', fontsize = 16)
plt.title('Dasies state space', fontsize = 16)
plt.show()
```

Now let's consider a case with green house effect, but without absorption of gases, setting $a= 10^{-4}$ and b= $0$

```python
model = Greenhouse_Daisywolrd( a=10e-4, b=0, area_b=0.01, area_w=0.01)
model.solve_model(store_values = True, iterations = 10000)
model.plot_model()
```

In this case we see that the greenhouse effect produces and abrupt raise in the temperature of the planet $Te$, and as a result the daises disappear form the planet. First, the increase in the population of white daisies provokes a decrease in the global temperature, but this triggers the increase of the population of black daisies, which in turn causes an increase in temperature that the black daisies can not bear. Then the white daisies grow again and this makes the temperature decrease a little, but the greenhouse effect moves the temperature away from the habitable zone, so the daisies go extinct. 

We can see how the areas of the daisies and the greenhouse effect interact making a 3d plot of these variables, and a 3d plot of the projections of this curve in each axis.

```python
model.plot_state_space()
```

In the above figure we can see the dynamics of the state space for the daises areas and the greenhouse effect. As $E$ increases, the populations oscillate and then they go extinct. Now let's see a case with both emission and absorption of green house gases setting $a=2.5^{-4}$ and $b=9^{-4}$

```python
model = Greenhouse_Daisywolrd( a=2.5e-4, b=9e-4, area_w = 0.01, area_b= 0.01)
model.solve_model(store_values = True, iterations = 50000)
model.plot_model()
model.plot_state_space(zlim=(0,0.6))
```

In this case the daisies can regulate the greenhouse effect for some time, but then the greenhouse effect overtakes the daisies and they go extinct. One way to fix this is increasing the absorption parameter $b$. In this way the greenhouse effect will be less harsh, and the daisies will be able to manage it. Interestingly, when we set $a=2.5^{-4}$ and $b= 11^{-4}$, we obtain a steady state in which the greenhouse effect is regulated and the withe daisies dominate. 

```python
model = Greenhouse_Daisywolrd( a=2.5e-4, b=11e-4, area_w = 0.01, area_b= 0.01)
model.solve_model(store_values = True, iterations = 100000)
model.plot_model()
model.plot_state_space(zlim=(0, 0.4))
```

If we continue increasing the $b$ parameter we obtain a situation in which there are self-sustained oscillations in the populations of daisies, and the greenhouse effect is regulated.

```python
model = Greenhouse_Daisywolrd( a=2.5e-4, b=10e-3, area_w = 0.3, area_b= 0.3)
model.solve_model(store_values = True, iterations = 10000)
model.plot_model(limits = False)
model.plot_state_space(zlim=(0,0.1))
```

Now, it would be interesting to see what happens when we change the luminosity of the sun. In the cell below we can see a plot of the change in areas and temperature with luminosity. In general, although the temperature of Daisyworld oscillates, it remains in a range comfortable for the daisies for values between 1.2 and 1.6. Now, since the daises oscillate with time, we would obtain different results in different stages of the oscillations, so it would be more appropriate to take  average values of the daisies areas, and of the temperatures, for different times.

```python
L = np.arange(0.5, 1.5, 0.05)
model = Greenhouse_Daisywolrd(a=2.5e-4, b=10e-3, area_w = 0.01, area_b= 0.01)
model.change_luminosity(L, iterations = 12000)
model.plot_change_luminosity()
```

In the next cell we can see the average areas of the daisies as the luminosity increases. This plot shows a similar pattern to the one in the original model. The mean, max and min temperatures are also shown. We can see that the mean global temperature is regulated between 1.6 and 1.2 $L$. Thus, when $a=2.5^{-4}$ and $b=10{-3}$, there is self regulation of the temperature on Daisyworld despite the green house effect. However, if we increase the emission parameter, or decrease the absorption parameter, this self regulation can disappear.

```python
L = np.arange(0.5, 1.5, 0.1)
model = Greenhouse_Daisywolrd(a=2.5e-4, b=10e-3, area_w = 0.01, area_b= 0.01)
model.plot_average_change_luminosity(L)
```

# Linear increase in luminosity

Paiva et.al. also make experiments in which the solar luminosity does not remain constant, but changes over time. However, rather than looking at the average behavior of the daisies for different luminosities as we did before, they modela linear increase in luminosity with the following equation: $L = 1.9x10^{-4}t + 0.75$. Thus, as time passes, the luminosity will increase, changing the behavior of the daisies. 

First, they consider a case where $a=b=0$:

```python
iterations = 50000
h= 0.1
t = np.arange(0, iterations*h, h) 
L_linear = 1.9e-4 * t +0.75
model = Greenhouse_Daisywolrd( a=0, b=0)
model.solve_model(store_values = True, iterations = iterations, h= h, L=L_linear)
model.plot_model()
```

In this case, the linear increase in luminosity makes the the temperature run away, so the daisies disappear from the planet. Now let's see what happens when $a=1^{-4}$ and $b = 0 $

```python
model = Greenhouse_Daisywolrd(a=1e-4, b = 0)
model.solve_model(store_values = True, iterations = iterations,  L=L_linear, h=h)
model.plot_model()
model.plot_state_space(zlim = (0,0.5))
```

In this case the daisies disappear from the planet in a much shorter time because of the combined effect of the linear increase in luminosity and of the greenhouse gases. Finally, let's make  $a=1^{-4}$ and $b=1^{-1}$

```python
model = Greenhouse_Daisywolrd(a=1e-4, b = 1e-1, area_b=0.1, area_w=0.1)
model.solve_model(store_values = True, iterations = iterations, L=L_linear)
model.plot_model(limits = False)
model.plot_state_space(zlim=(0,0.015))
```

In this case the populations oscillate and regulate the greenhouse effect for sometime. However, the linear increase in luminosity makes grow the temperature until a point that the daisies can not bear. AT this point the greenhouse effect in triggered. 

# Luminosity representing Climate Variability

Paiva et. al. also make experiments with a more realistic luminosity that represents climate variability. To do this, they set the luminosity to the following function: $L = 1.9x10^{-4}t +0.75 + N \sin (w t)$, where $t$ is the time, $N$ is the amplitude of the oscillations, and $w$ is their frequency. Here we will set  $N=0.1$ and $w = 0.02$. Let's plot this function


```python
iterations = 60000
h= 0.1
t = np.arange(0, iterations*h, h) 
L_linear = 1.9e-4 * t +0.75
N = 0.1
w = 0.02
Solar_Lum = L_linear + N*np.sin(w*t) 

plt.plot(t, Solar_Lum)
plt.xlabel('time (days)')
plt.ylabel('Solar Luminosity L');
```

We see that now the luminosity oscillates while it increases linearly. Let's see what hapens in Daisyworld in this situation. First, let's set $a=b =0$.

```python
model = Greenhouse_Daisywolrd( a=0, b=0)
model.solve_model(store_values = True, iterations = iterations, L=Solar_Lum)
model.plot_model(limits = False)
```

When there is no absorption nor production of greenhouse gases, the daisies populations, and the temperature, oscillate for some time. However, the increase in luminosity makes increase the temperature to a point that the daisies can not bear. It is interesting to note that the temperature and the luminosity have more oscillations in the case in comparison with the linear luminosity case. Now, let's set $a=1^{-4}$ and $b=0$.

```python
model = Greenhouse_Daisywolrd( a=1e-4, b=0)
model.solve_model(store_values = True, iterations = iterations, L=Solar_Lum, h=h)
model.plot_model()
model.plot_state_space()
```

In this case the combined effect of the increased luminosity and the greenhouse effect makes the daisies disappear in a shorter time. Finally, let's set $a=1^{-4}$ and $b= 1^{-1}$

```python
model = Greenhouse_Daisywolrd( a=1e-4, b=1e-1)
model.solve_model(store_values = True, iterations = iterations, L=Solar_Lum, h=h)
model.plot_model(limits = False)
model.plot_state_space(zlim = (0,0.1))
```

In this case the daisies can regulate the greenhouse effect for about 4000 days. Then the daisies go extinct and the greenhouse effect starts to grow.

# Local Analysis

Finally, Paiva et. al. perform a local analysis of Daisyworld with greenhouse gases, looking at the solar energy input in a single point of the planet. They propose to make the analysis at the equator, where the solar luminosity in zero in the nights, and changes form zero to its maximum value during the day. The authors propose to use a function for the local solar luminosity to represent this daily and seasonal variability. Here, i will be using a variation of their original function:

\begin{equation*}
L = [M1 \sin (w t) + M2)]  \sin (2 \pi t)
\end{equation*}

WHERE $M1 = 0.1 $ and $M2 = 1$ are amplitudes, $w = 0.0175$ corresponds to a period of one year, and $t$ is the time in days. We can break this function into two parts. The first one ($[M1 \sin (w t) + M2)]$) controls the seasonal cycle, and the second part ($\sin (2 \pi t$) controls the daily cycle. Let's plot this function to better understand it.

```python
def local_L(t):
    '''Calculates local solar luminosity'''
    L =  (0.1 * np.sin(0.0175 * t) + 1 ) * (np.sin(2* t * np.pi) )
    return L.clip(min = 0)

t = np.arange(0, 1000,  0.05)
L = local_L(t)

plt.figure(figsize=(12, 4))
plt.subplot(1,2,1)
plt.plot(t, L)
plt.xlabel('time (days)')
plt.ylabel('Luminosity')
plt.title('Local luminosity dynamics in 3 years')

plt.subplot(1,2,2)
plt.plot(t[:60], L[:60])
plt.xlabel('time (days)')
plt.ylabel('Luminosity')
plt.title('Local luminosity dynamics in 3 days');
```

We can see that that the local solar luminosity function represents a solar luminosity that varies seasonally through the year. Also, in the first half of a single day the solar luminosity varies form zero to its maximum, and it is zero in the second half of the day. This is a somewhat realistic representation of luminosity at the equator. Now, let's experiment with this local luminosity in Daisyworld!

```python
def local_L(t):
    '''Calculates local solar luminosity'''
    L =  (1.9 * np.sin(0.0175 * t) + 3.85 ) * (np.sin(2* t * np.pi) )
    return L.clip(min = 0)
```

```python
iterations = 700000
h = 0.01
tf = iterations * h
t = np.arange(0, tf,  h)
local_luminosity = local_L(t)
plt.plot(t, local_luminosity)
print(len(local_luminosity))
model = Greenhouse_Daisywolrd( a=1e-5, b=10e-2)
model.solve_model(store_values = True, iterations = iterations, L=local_luminosity, h=h)
model.plot_model(limits = False)
```
